```python
# Import libraries
from bs4 import BeautifulSoup
import requests
import time # timer
import datetime # get timestamp
import csv
import pandas as pd
import os.path
import smtplib # sending emails


def check_price():
    
    # STEP 1: Connect to the website (Amazon)
    url = 'https://www.amazon.es/Apple-MWP22TY-A-AirPods-Pro/dp/B07ZPNLGDP/ref=sr_1_3_sspa?__mk_es_ES=%C3%85M%C3%85%C5%BD%C3%95%C3%91&dchild=1&keywords=earpods&qid=1631817859&sr=8-3-spons&psc=1&spLa=ZW5jcnlwdGVkUXVhbGlmaWVyPUFEWUYyNDlLUlhLUFYmZW5jcnlwdGVkSWQ9QTA4MDkxNjYySlc0NUVUM0pQVjBGJmVuY3J5cHRlZEFkSWQ9QTA1MTE1NzIyT1RTMUJNUjdCRFFDJndpZGdldE5hbWU9c3BfYXRmJmFjdGlvbj1jbGlja1JlZGlyZWN0JmRvTm90TG9nQ2xpY2s9dHJ1ZQ=='
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
    # Tip for headers -> Get User-Agent via this url: http://httpbin.org/get


    # STEP 2: Getting the HTML of the page
    page = requests.get(url, headers = headers)
    soup1 = BeautifulSoup(page.content, "html.parser")
    soup2 = BeautifulSoup(soup1.prettify(), "html.parser") # to better formatting the html code, just in case :)


    # STEP 3: Getting the content that we want from the page
    # 1. Title of the product
    title = soup2.find(id="productTitle").get_text()
    # 2. Price of the product
    price = soup2.find(id="priceblock_ourprice").get_text()
    # 3. Rating of the product (pending)
    # rating = soup2.find().get()


    # STEP 4: Cleaning the data we got and adding useful data
    # price: eliminate spaces and get only the number
    price = price.strip().replace(",", ".")[:6]
    # title: eliminate spaces
    title = title.strip()
    #create timestamp to add to the data
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") #strftime for formatting


    # STEP 5: Create csv to push into the data and append new data to it
    # create the headers of the csv file
    head = ['date', 'title', 'price']
    # create a list including the data we got
    data = [timestamp, title, price]


    #loop for creating the csv and appending data to it
    if os.path.exists(r'C:\Users\Adrián\Desktop\Main\DataProjects\WebScraper\AmazonWebScraperDB.csv'): #checking if the file already exists to create it or append
        with open("AmazonWebScraperDB.csv", 'a+', newline='', encoding='UTF8') as f: #a+ to append data
            writer = csv.writer(f)
            writer.writerow(data)
    else:
        # if the file doesnt exist, creating the csv and then inserting the header and the data (for the first time)
        with open("AmazonWebScraperDB.csv", 'w', newline='', encoding='UTF8') as f:
            writer = csv.writer(f)
            writer.writerow(head)
            writer.writerow(data)
            
            
    #STEP 6: Send an alert to your email if price drops below your limit
    limit = 150 # set the max amount you want to pay for the product
    if float(price)<limit:
        # configure SMTP Connection
        server = smtplib.SMTP('smtp.office365.com',587)
        server.ehlo()
        server.starttls()
        server.ehlo()
        server.login('me@hotmail.com','notapassword')      
        # edit the message
        subject = "Airpods Pro's just dropped below {}!".format(limit)
        body = """Hey! Airpods Pro's price is {}euros. Time to buy them!
        \nLink here: {}
        """.format(price, url)
        msg = 'Subject: {}\n\n{}'.format(subject, body)
        # send email (sender, receiver, message)
        server.sendmail('me@hotmail.com', ['me@hotmail.com'], msg)

#check_price()
# Run the code automatically every X seconds
while(True):
    frequency = 30 # set the frequency (in seconds) to run the code
    check_price()
    time.sleep(frequency)

```
